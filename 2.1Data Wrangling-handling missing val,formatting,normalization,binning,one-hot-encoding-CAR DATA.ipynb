## rectifying the data formats

```python
#replacing normalized-losses column missing values with mean but before replacing the data type must be correct

'''
The following have the wrong data type of OBJECT
    normalized-losses
    bore
    stroke
    horsepower
    peak-rpm
    price
'''
df.dtypes
```

```python
#Rectifying the data types
df[["normalized-losses"]]=df[["normalized-losses"]].astype('float')
df[["bore"]]=df[["bore"]].astype('float')
df[["stroke"]]=df[["stroke"]].astype('float')
df[["horsepower"]]=df[["horsepower"]].astype('float')
df[["peak-rpm"]]=df[["peak-rpm"]].astype('float')
df[["price"]]=df[["price"]].astype('float')
```

```python
df.head()
```

## Inserting missing values with mean

```python
#replacing normalized-losses column missing values with mean


df['normalized-losses'].replace(np.nan,df['normalized-losses'].mean(),inplace=True)
df['bore'].replace(np.nan,df['bore'].mean(),inplace=True)
df['stroke'].replace(np.nan,df['stroke'].mean(),inplace=True)
df['horsepower'].replace(np.nan,df['horsepower'].mean(),inplace=True)
df['peak-rpm'].replace(np.nan,df['peak-rpm'].mean(),inplace=True)


```

```python
df.head(30)
#Null values replaced with mean values
```

```python
#value_counts() tells : The values present in a column+how many times each value occured
df['num-of-doors'].value_counts()
```

```python
#value_counts().idmax() tells: The values present in the column +which value occured most number of times in that column
df['num-of-doors'].value_counts().idxmax()
#using this,we can replace missing values in a column with maximum occuring value

df['num-of-doors'].replace(np.nan,'four',inplace=True)#because most seadns have 4 doors
```

```python
#dropping the row that does not even have a price
df.dropna(subset=['price'],axis=0,inplace=True)


```

```python
#after droppping a row,reset the index:
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

## Data standardization

#### the fuel consumption columns "city-mpg" and "highway-mpg" are represented by mpg (miles per gallon) unit. 
#### convert it to L/100km standard

```python
df['city-mpg'].head()

df['city-L/100km']=235/df['city-mpg']#creating a new column df['city-L/100km]
df['city-L/100km'].head()
```

```python
df['highway-mpg'].head()

df['highway-L/100km']=235/df['city-mpg']#creating a new column df['highway-L/100km]
df['highway-L/100km'].head()

```

## Normalizing length width and height columns

```python
#normalizing using standard scaling
df['length']=df['length']/df['length'].max()
df['width']=df['width']/df['width'].max()
df['height']=df['height']/df['height'].max()
```

## Binning

```python
#binning the horsepower variable into -low,medium and high Bins

bins=np.linspace(min(df['horsepower']),max(df['horsepower']),4)#we need 4 dividers for 3 categories of bins

bin_names=['low','medium','high']

df['horsepower_binned']=pd.cut(df['horsepower'],bins,labels=bin_names,include_lowest=True)#cut the horsepower column into bins with label name=bin_names
```

```python
df['horsepower_binned'].value_counts()
```

```python
# plotting Bins

plt.bar(bin_names,df['horsepower_binned'].value_counts())
plt.xlabel('names')
plt.ylabel('count')
plt.title('Horsepower Bins')
plt.show()

#shows-Maximum cars have low horse power
```

```python
plt.hist(df["horsepower"], bins = 3)
#same output as above
```

## Creating dummy variable dataframe and then merging with the original data frame

#### fuel_type variable has either gas or diesel and we will make dummy values for it

```python
dummy_values=pd.get_dummies(df['fuel-type'],dtype=int)#if we dont write dtype=int then it will give true/false as dummy values instead of 1/0
dummy_values.head()
```

```python
#changing names of dummy columns
dummy_values.rename(columns={'diesel':'fuel-type-diesel','gas':'fuel-type-gas'},inplace=True)
```

```python
dummy_values.head()
```

```python
#concatenating the two dataframes

pd.concat([df,dummy_values],axis=1)
```

```python
#dropping the original fuel-type column

df.drop("fuel-type", axis = 1, inplace=True)#Note-The syntax is tricky in dropping
```
