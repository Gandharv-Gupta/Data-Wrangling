## handling missing values

```python
null_table=df.isnull()
null_table.head()
```

```python
#counting null values
for col in null_table.columns.values.tolist():
    print(col)
    print(null_table[col].value_counts())
    print("")
```

## rectifying datatypes before replacing null values

```python
df.dtypes
```

```python
df.head()
```

```python
#all datatypes are correct
```

## replacing weight  null values with mean weight

```python


df['Weight_kg']=df['Weight_kg'].replace(np.nan,df['Weight_kg'].mean())
```

## screen_size_cm is a categorical variable and hence we will replace the missing values with most occuring value



```python
df['Screen_Size_cm'].value_counts().idxmax()
```

```python
df['Screen_Size_cm']=df['Screen_Size_cm'].replace(np.nan,39.624)
```

## standardize weight in pounds and screen_size_cm in inches

```python
#   1 inch = 2.54 cm
#   1 kg   = 2.205 pounds
```

```python
#screen_size:
df['Screen_Size_cm']=df['Screen_Size_cm']/2.54

#weight:
df['Weight_kg']=df['Weight_kg']*2.205
```

## Normalizing CPU_frequency with its maximum value-standard scaling

```python
df['CPU_frequency']=df['CPU_frequency']/df['CPU_frequency'].max()
df['CPU_frequency'].head()
```

## Binning price into high medium and low

```python
# using 4 dividers for 3 bins
bins=np.linspace(min(df['Price']),max(df['Price']),4)# note-min must always come before max ,not after in the syntax

bin_name=['low','medium','high']

df['Price_binned']=pd.cut(df['Price'],bins,labels=bin_name,include_lowest=True)

df['Price_binned'].head()
```

```python
## Visualizing the distribution of bins:

plt.bar(bin_name,df['Price_binned'].value_counts())
plt.xlabel('bins')
plt.ylabel('count')
plt.show()

#shows -maxium laptops have low price
```

## Indicator variables/Categorical to quantitative variables:(for screen variable)

```python
df['Screen'].value_counts()
```

```python
#creating dummies for full hd and ips panel

dummy_table=pd.get_dummies(df['Screen'],dtype=int)
dummy_table.head()
```

```python
#concatenating the original df and dummy table

df=pd.concat([df,dummy_table],axis=1)
```

```python
df.head()
```

```python
#changing column names that were added
df.columns

df.rename(columns={'Full HD':'screen_Full_HD','IPS Panel':'screen_IPS_Panel'},inplace=True)
```

```python
df.head()
```

```python
#dropping the original screen column as we already added dummy columns

df.drop('Screen',axis=1,inplace=True)
```

```python
## data is cleaned and can be exported inn csv format:

df.to_csv(r"C:\Users\HP\Desktop\CODE\DATA ANALYSIS with PYTHON\cleaned_data_for2.2.csv",index=False)
```
